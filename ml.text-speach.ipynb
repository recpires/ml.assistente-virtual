- Importando Biblioteca


```python
from gtts import gTTS

text_to_say = "But seek first the kingdom of God and his righteousness, and all these things will be added to you. Mateus 6:33"

language = "en"

gtts_object = gTTS(text = text_to_say,
                   lang = language,
                   slow = False)

gtts_object.save("/content/audio/gtts.wav")
```
